# IPL Dataset : 2008 to 2017

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
ipl = pd.read_csv("matches.csv")
```

```python
ipl.head()
```

```python
ipl.info()
```

```python
ipl.drop(columns=['umpire3'], inplace = True)
```

```python
ipl.team1.value_counts()
```

```python
ipl['team1'] = ipl['team1'].map({'Sunrisers Hyderabad': 'SRH', 'Mumbai Indians': 'MI', 'Chennai Super Kings': 'CSK','Royal Challengers Bangalore':'RCB', 'Kolkata Knight Riders' : 'KKR','Delhi Daredevils':'DD', 'Rajasthan Royals': 'RR', 'Deccan Chargers':'DC', 'Pune Warriors': 'PW','Gujarat Lions': 'GL','Rising Pune Supergiant': 'RPS','Rising Pune Supergiants': 'RPS', 'Kochi Tuskers Kerala': 'KT', 'Kings XI Punjab': 'KXIP'})

ipl['team2'] = ipl['team2'].map({'Sunrisers Hyderabad': 'SRH', 'Mumbai Indians': 'MI', 'Chennai Super Kings': 'CSK','Royal Challengers Bangalore':'RCB', 'Kolkata Knight Riders' : 'KKR','Delhi Daredevils':'DD', 'Rajasthan Royals': 'RR', 'Deccan Chargers':'DC', 'Pune Warriors': 'PW','Gujarat Lions': 'GL','Rising Pune Supergiant': 'RPS','Rising Pune Supergiants': 'RPS', 'Kochi Tuskers Kerala': 'KT', 'Kings XI Punjab': 'KXIP'})

ipl['toss_winner'] = ipl['toss_winner'].map({'Sunrisers Hyderabad': 'SRH', 'Mumbai Indians': 'MI', 'Chennai Super Kings': 'CSK','Royal Challengers Bangalore':'RCB', 'Kolkata Knight Riders' : 'KKR','Delhi Daredevils':'DD', 'Rajasthan Royals': 'RR', 'Deccan Chargers':'DC', 'Pune Warriors': 'PW','Gujarat Lions': 'GL','Rising Pune Supergiant': 'RPS','Rising Pune Supergiants': 'RPS', 'Kochi Tuskers Kerala': 'KT', 'Kings XI Punjab': 'KXIP'})

ipl['winner'] = ipl['winner'].map({'Sunrisers Hyderabad': 'SRH', 'Mumbai Indians': 'MI', 'Chennai Super Kings': 'CSK','Royal Challengers Bangalore':'RCB', 'Kolkata Knight Riders' : 'KKR','Delhi Daredevils':'DD', 'Rajasthan Royals': 'RR', 'Deccan Chargers':'DC', 'Pune Warriors': 'PW','Gujarat Lions': 'GL','Rising Pune Supergiant': 'RPS','Rising Pune Supergiants': 'RPS', 'Kochi Tuskers Kerala': 'KT', 'Kings XI Punjab': 'KXIP'})
```

```python
ipl['date'] = pd.to_datetime(ipl.date)

ipl.head()
```

```python
plt.figure(figsize=(15,6))
ax = sns.countplot(x="season", data=ipl, palette="viridis")
plt.title('Number of games played in each season', fontsize= 25, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="muted")
ipl.dl_applied.value_counts().plot.pie(autopct='%1.0f%%', pctdistance=0.7, labeldistance=1.08)
plt.axis('equal')
plt.title('Total number of wins by a Team in all seasons', fontsize= 25, fontweight="bold")
pylab.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="colorblind")
ipl.winner.value_counts().plot.bar()
plt.title('Total number of wins by a Team in all seasons', fontsize= 25, fontweight="bold")
plt.xlabel('Teams', fontsize=18, fontweight="bold")
plt.ylabel('Wins', fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="muted")
ipl.winner.value_counts().plot.pie(autopct='%1.0f%%', pctdistance=0.7, labeldistance=1.08)
plt.axis('equal')
plt.title('Total number of wins by a Team in all seasons', fontsize= 25, fontweight="bold")
pylab.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="rocket")
ipl.city.value_counts()[0:15,].plot.bar()
plt.title('Top 15 cities hosted IPL matches', fontsize= 25, fontweight="bold")
plt.xlabel('Cities', fontsize=18, fontweight="bold")
plt.ylabel('Matches hosted', fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="dark")
ipl.venue.value_counts()[0:15,].plot.bar()
plt.title('Top 15 stadiums hosted IPL matches', fontsize= 25, fontweight="bold")
plt.xlabel('Stadium', fontsize=18, fontweight="bold")
plt.ylabel('Matches hosted', fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="muted")
ipl[ipl['city'] == "Mumbai"]["venue"].value_counts().plot.bar()
plt.title("Mumbai has 3 stadiums. That's the reason why Wankhede stadium is not in top stadiums hosted IPL matches",fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="colorblind")
ipl[ipl['season'] == 2016]['winner'].value_counts().plot.bar()
plt.title('Number of wins by Team in 2016', fontsize= 25, fontweight="bold")
plt.xlabel('Teams', fontsize=18, fontweight="bold")
plt.ylabel('Wins', fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="rocket")
ipl.player_of_match.value_counts()[0:10,].plot.bar()
plt.title('Most Player of the match awards among all seasons', fontsize=25, fontweight="bold")
plt.xlabel('Players', fontsize=18, fontweight="bold")
plt.ylabel('Awards', fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.set(font_scale=1.5, palette="dark")
ipl.toss_decision.value_counts().plot.bar()
plt.title('Toss decision among all seasons', fontsize=25, fontweight="bold")
plt.xlabel('Decisions', fontsize=18, fontweight="bold")
plt.ylabel('No.of times', fontsize=20, fontweight="bold")
plt.show()
```
